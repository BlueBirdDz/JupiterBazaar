```python
# Est-ce que la pollution se concentre sur un établissement, ou est-elle équitablement répartie ? ( par polluant)
# Quels sont les principaux établissements responsables ?
# On affiche les courbes de Lorenz pour chaque polluant
for i in list(emissions.polluant.unique()):
    emissions_polluant=(emissions.loc[emissions['polluant']== i])
    pollution = emissions_polluant['quantite']
    n = len(pollution)
    lorenz = np.cumsum(np.sort(pollution)) / pollution.sum()
    lorenz = np.append([0],lorenz) # La courbe de Lorenz commence à 0

    plt.axes().axis('equal')
    xaxis = np.linspace(0-1/n,1+1/n,n+1) #Il y a un segment de taille n pour chaque individu, plus 1 segment supplémentaire d'ordonnée 0. Le premier segment commence à 0-1/n, et le dernier termine à 1+1/n.
    plt.plot(xaxis,lorenz,drawstyle='steps-post')
    plt.title(i)
    plt.show()
    print(emissions_polluant.sort_values(by=['quantite'], ascending=False).head())
    
```

```python
X='code_ape'
Y='quantite'

def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
eta_squared(ttt[X],ttt[Y])
```
